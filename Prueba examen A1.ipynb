Derivada $f'(x) = -\frac{\beta}{V^2}-\frac{2\gamma}{V^3}-\frac{3\delta}{V^4}-P$


## **Helio** 
\
Valor inicial: $V=\frac{RT}{P}=\frac{0.08205\times 423}{75}=0.462762$ \
Primer iteración \
$V = V_0 - \frac{f(V_0)}{f'(V_0)}=0.612093-\frac{RT - P \times 0.612093 + \frac{\beta}{(0.612093)}+\frac{\gamma}{(0.612093)^2}+\frac{\delta}{(0.612093)^3}}{-\frac{\beta}{(0.612093)^2}-\frac{2\gamma}{(0.612093)^3}-\frac{3\delta}{(0.612093)^4}-50}=  \\
V = 0.612093-\frac{0.08205 \times 373.15 - 50 \times 0.612093 + \frac{0.407037405}{(0.612093)}+\frac{0.001292544}{(0.612093)^2}+\frac{0}{(0.612093)^3}}{-\frac{0.407037405}{(0.612093)^2}-\frac{2\times 0.001292544}{(0.612093)^3}-\frac{3\times 0}{(0.612093)^4}-50}= 0.6254155217346812$

```python
V = 0.462762
P = 75
T = 423.15
A0 = 0.0216
a = 0.05984
B0 = 0.014
b = 0
c = 0.00000040
R = 0.08205746
beta = R*T*B0-A0-((R*c)/(T**2))
gamma = -R*T*B0*b+A0*a-((R*B0*c)/(T**2))
delta = (R*B0*b*c)/(T**2)
X = Symbol('X')
funcion = R*T-P*X+(beta/X)+(gamma/(X**2))+(delta/(X**3))
diff(Symbol('R')*Symbol('T')-Symbol('P')*X+(Symbol('beta')/X)+(Symbol('gamma')/(X**2))+(Symbol('delta')/(X**3)),X)
```

```python
MN_NR(funcion, V)
```
