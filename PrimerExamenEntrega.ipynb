#### Ejercicio 1: Se implementa el método de Newton-Raphson para la ecuación de Redlich-Kwong
<center> $ab = PT^{0.5}V^3-RT^{1.5}V^2-(Pb^2T^{0.5}-a+RT^{1.5}b)V$ </center>

```python
# Creamos una función que realice el Método de Newton - Raphson para 
# la función de Redlich-Kwong ab = PT^(0.5)V^(3)-RT^(1.5)V^(2)-(Pb^(2)T^(0.5)-a+RT^(1.5)b)V

def Newton_Raphson(x_0,P,R,T,a,b): # Metemos las magnitudes que intervienen en la ecuación
    gxs = []
    results = [] 
    fin = 1
    while fin == 1:
        # Configuramos la operación iterativa que representada como x_(i+1) = f(x_i)/f'(x_i) 
        # Función -> f(x) PT^(0.5)V^(3)-RT^(1.5)V^(2)-(Pb^(2)T^(0.5)-a+RT^(1.5)b)V-ab = 0
        # Derivada dF/dV -> f'(x) = 3PT^(0.5)V^(2)-2RT^(1.5)V-(Pb^(2)T^(0.5)-a+RT^(1.5)b)
        x_0 = x_0 - ((P*(T**0.5)*x_0**3)-(R*(T**1.5)*x_0**2)-(((P*b**2*(T**0.5))-(a)+(R*(T**1.5)*b))*x_0)-a*b) / ((3*P*(T**0.5)*x_0**2)-(2*R*(T**1.5)*x_0)-((P*b**2*(T**0.5))-(a)+(R*(T**1.5)*b))) 
        # Se evalúa el valor recien calculado en la función original, para verificar la aproximación
        resu =  ((P*(T**0.5)*x_0**3)-(R*(T**1.5)*x_0**2)-(((P*b**2*(T**0.5))-(a)+(R*(T**1.5)*b))*x_0)-a*b)
        print(x_0)
        gxs.append(x_0)
        results.append(resu) 
        # Se evalúa la precisión de la aproximación para detener la ejecución.
        if -0.000000001 < resu < 0.000000001:
            fin = 0 

    allres = { 'x_i': gxs,
                'f(x)': results}

    allresdf = pd.DataFrame(allres)
    return allresdf
```

```python
# Estabecemos las magnitudes generales dadas, y las específicas para el Helio

#Presión en atm
P = 50
# Temperatura en Kelvins
T = 373.15
# Temperatura crítica del Helio en Kelvins
Tc = 5.26 
#Presión crítica en atms
Pc = 2.26
# Constante ideal de los gases en (atm L)/(gmol K)
R = 0.08205
#Volumen molar inicial en L/gmol 
V = (R*T)/P
x_0 = V
# Calculamos a
a = 0.4278*((R**2*(Tc**2.5))/(Pc))
#Calculamos b
b = 0.0867*((R*Tc)/(Pc))

# Llamamos a la función para correr el métos Newton-Raphson sobre la función de Redlich-Kwong
Newton_Raphson(x_0,P,R,T,a,b)
```

#### Ejercicio 2

Para la función 

<center> $x^3-6x+2=0$ </center> 

Corremos tres depejes distintos y vemos con cual se encuentra una raíz entre [2,3].

######  Con el despeje a) $x = \sqrt[3]{6x-2} $

```python
# Elegimos un valor intermedio como nuestro valor inicial dados los parametros [2,3]
res = 2.5

fin = 1
gxs = []
results = []

while fin == 1:
   # Evaluamos el valor inicial en el depeje a)
    res = np.cbrt(6*res-2)
    # Evaluamos el valor recien calculado en la función original
    resu = res**3-6*res+2
    gxs.append(res)
    results.append(resu) 
    # Establecemos un criterio de paro del calculo, con el resultado de la evaluación 
    # del valor en la función original 
    if -0.00001 < resu < 0.00001:
        fin =0
    
allres = { 'g(x)': gxs,
           'f(x)': results}

allresdf = pd.DataFrame(allres)
allresdf
```

##### Ejercicio 3: Implementamos el método de Newton-Raphson Mutivariable para el sistemad de ecuaciones:

<center> $ln(xy)+x^2y^2=8$ </center>
<center> $sen(x)+y\; e^x=2$ </center>

```python
#Proponiendo valores iniciales 
x = 100
y = 100

fin = 1
xs = []
ys = []
cambiosx = []
cambiosy = []
cont = 0

while fin == 1:
    cont += 1
    
    x_anterior = x
    y_anterior = y
    
    # Función 1  ln(xy) + x^2 y^2 - 8 = 0
    f1 = m.log(x*y)+(x**2)*(y**2)-8
    #Función 2 sen(x) + y e^x - 2 = 0
    f2 = m.sin(mpmath.radians(x)) +y*m.exp(x) -2
    
    # Obtenemos la derivada df1/dx = 1/x + 2x y^2
    df1dx = (1/x)+2*x*(y**2)
    # Obtenemos la derivada df1/dy = 1/y +2y x^2
    df1dy = (1/y)+2*y*(x**2)
    # Obtenemos la derivada df2/dx = cos(x)+y e^x7
    df2dx = m.cos(mpmath.radians(x))+y*m.exp(x)
    # Obtenemos la derivada df2/dy
    df2dy = m.exp(x)
    # Creamos la Matriz Jacobiana y sacamos su inversa
    jac_inv = np.linalg.inv([[df1dx,df1dy], [df2dx,df2dy]])
    primat = Matrix(jac_inv)
    # Se multiplica (J(x,y)^-1)([[-f1],[-f2]])
    h = jac_inv.dot([-f1,-f2])
#     h = np.array([(-f1)*m_i[0,0]+(-f1)*m_i[1,0],(-f2)*m_i[0,1]+(-f2)*m_i[1,1]])
    
    # Recalculamos los valores de 'x' y 'y' 
    x += h[0]
    y += h[1]
    
    xs.append(x)
    ys.append(x)
    
    # Medimos el cambio respecto a la iteración anterior
    cambiox = x_anterior - x
    cambioy = y_anterior - y
    
    cambiosx.append(cambiox)
    cambiosy.append(cambioy)
    
    print(cont,'\t x: {0:.6f}'.format(x), ' y: {0:.6f}'.format(y) ,' cambio en x: {0:.6f}'.format(cambiox), ' cambio en y: {0:.6f}'.format(cambioy))
    
    # Si el cambio es lo suficientemente pequeño se detiene el proceso.
    if 0.0000001 < cambiox < 0.0000001 and 0.0000001 < cambioy < 0.0000001:
        fin = 0
    
# primat
```
